```python
#比較統計學的同學 玩手機遊戲的時間跟網路流量是不是吃到飽的關係
#因為調查是匿名的 所以每一位同學都只會有編號
import pandas as pd

#讀取csv的檔案
stat = pd.read_csv ("stat.csv")
```

```python
#列出手機調查資料
print(stat)
```

```python
#顯示調查的樣本數
data = list(stat['Data'])
time = list(stat['Time'])
print("手機調查樣本數:", len(data))
```

```python
#列出集合
set_data = set(stat['Data'])
set_time = set(stat['Time'])
set_data.remove('FALSE')
set_time.remove('FALSE')
print(set_data)
print(set_time)
```

```python
#列出交集
set_data & set_time
print(set_data & set_time)
print("共", len(set_data & set_time), "比資料")

#重複的資料偏多 固有一定的關聯性
```

```python
#列出聯集
set_data | set_time
print(set_data | set_time)
print("共", len(set_data | set_time), "比資料")
```

```python
#列出差集
print(set_data - set_time)
print("共", len(set_data - set_time), "比資料")
print(set_time - set_data)
print("共", len(set_time - set_data), "比資料")
```

```python
#列出對稱差集
set_a = set_data | set_time
set_b = set_data & set_time
print(set_a-set_b)
print("共", len(set_a-set_b), "比資料")
```
